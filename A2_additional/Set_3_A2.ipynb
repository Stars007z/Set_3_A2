```python
!git clone https://github.com/Stars007z/Set_3_A2.git

import matplotlib.pyplot as plt

def read_data(filename):

    sizes = []
    trand = []
    treverse = []
    tnear = []

    with open(filename, 'r') as file:
        lines = file.readlines()
        for line in lines[1:]:
            size, rand_time, reverse_time, near_time = map(int, line.split())
            sizes.append(size)
            trand.append(rand_time)
            treverse.append(reverse_time)
            tnear.append(near_time)

    return sizes, trand, treverse, tnear


def plot_sorting_results(file, algorithm_name):
    sizes, trand, treverse, tnear = read_data(file)

    plt.figure(figsize=(12, 8))
    plt.plot(sizes, trand, marker='o', label="Random Array")
    plt.plot(sizes, treverse, marker='s', label="Reverse Array")
    plt.plot(sizes, tnear, marker='^', label="Nearly Sorted Array")

    plt.title(f"{algorithm_name} Sorting Performance", fontsize=16)
    plt.xlabel("Array Size (n)", fontsize=14)
    plt.ylabel("Time (microsec)", fontsize=14)
    plt.grid(True)
    plt.legend(fontsize=12)
    plt.show()


print("График для Merge Sort:")
plot_sorting_results("Set_3_A2/A2_additional/ms_results.txt", "Merge Sort")
print("График для Hybrid Merge-Insert Sort:")
plot_sorting_results("Set_3_A2/A2_additional/hms_results.txt", "Hybrid Merge-Insert Sort")
```
